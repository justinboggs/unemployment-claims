```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.dates as dates
```

```python
data = pd.read_csv('data/UNRATE.csv')
```

```python
data.head()
```

```python
unemploy_rate = data.rename(columns={'DATE': 'Date', 'UNRATE': 'Rate'})
unemploy_rate.head()
```

```python
x_axis = unemploy_rate['Date']
y_axis = unemploy_rate['Rate']
```

```python
fig = plt.figure(figsize=(40,15))
ax = fig.add_subplot()
chart = plt.plot(unemploy_rate['Date'], unemploy_rate['Rate'], color='blue', alpha=0.9)
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
plt.xticks(rotation=90)
plt.savefig('output/monthly_rate.png')
plt.show()
```
