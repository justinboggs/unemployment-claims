```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.dates as dates
import io
import requests
```

```python
unemploy_data = pd.read_csv("data/ICNSA.csv")
```

```python
unemploy_data.tail()
```

```python
initial_claims = unemploy_data.rename(columns={'DATE':'Date', 'ICNSA':'Claims'})
initial_claims.tail()
```

```python
initial_claims['Year'] = pd.DatetimeIndex(initial_claims['Date']).year
initial_claims.tail()
```

```python
x_axis = initial_claims['Date']
y_axis = initial_claims['Claims']
year = initial_claims['Year']
```

```python
fig = plt.figure(figsize=(40,15))
ax = fig.add_subplot()
chart = plt.plot(initial_claims['Date'], initial_claims['Claims'], color='blue', alpha=0.9)
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
plt.xticks(rotation=90)
plt.ylabel('# of Initial Claims')
plt.title('Unemployment Claims 1967 - Present')
plt.savefig('output/weekly_claims.png')
plt.show()
```
